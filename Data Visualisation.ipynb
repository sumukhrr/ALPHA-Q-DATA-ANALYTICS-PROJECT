```python
"""Correlation map is defined here
It gives correlation between the main columns on which the price depends the most that is Year and Odometer"""
def category_values(dataframe, categories):
    for c in categories:
        print('\n', dataframe.groupby(by=c)[c].count().sort_values(ascending=False))
        print('Nulls: ', dataframe[c].isnull().sum())

def plot_correlation_map( df ):
    corr = df.corr()
    _ , ax = plt.subplots( figsize =( 10 , 8 ) )
    cmap = sns.diverging_palette( 220 , 9 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
```

```python
"""As we can see Year and price are positively correlated to each other where as
Odometer and Price is negatively correlated to each other.
And the surprising relation is even Year and Odometer are correlated to each other negatively"""
plot_correlation_map(df)
df.corr()
```
